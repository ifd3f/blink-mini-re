```python
import pandas as pd

def read_objdump(f):
    # skip first 7 lines
    for i in range(7):
        f.readline()

    for line in f:
        inst = line.strip().split('\t')
        addr, data, op, arg, *comments = inst + [''] * (6 - len(inst)) # pad with blank
        comment = ''.join(comments)

        yield int(addr[:-1], 16), int(data, 16), op, arg, comment


with open('./notes/objdump.txt') as f:
    df = pd.DataFrame(read_objdump(f), columns=['addr', 'data', 'op', 'arg', 'comment'])
    df
```

```python
bad_insts = df['comment'].str.contains('UNDEFINED') | df['comment'].str.contains('UNPREDICTABLE')
df = df.loc[~bad_insts]
df
```
